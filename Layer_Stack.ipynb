## **Import Libraries**

```python
import rasterio
from rasterio.plot import show
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Open Bands
nir_bands_path = "/content/drive/MyDrive/Machine Learning Dataset/Sentinal/B8.tif"
red_bands_path = "/content/drive/MyDrive/Machine Learning Dataset/Sentinal/B4.tif"
green_bands_path = "/content/drive/MyDrive/Machine Learning Dataset/Sentinal/B3.tif"
blue_bands_path = "/content/drive/MyDrive/Machine Learning Dataset/Sentinal/B2.tif"
```

```python
# Open each band
with rasterio.open(nir_bands_path) as src:
    nir_bands = src.read(1)

with rasterio.open(red_bands_path) as src:
    red_bands = src.read(1)

with rasterio.open(green_bands_path) as src:
    green_bands = src.read(1)

with rasterio.open(blue_bands_path) as src:
    blue_bands = src.read(1)
```

```python
# Convert the bands to float32 to avoid overflow issues
nir_bands = nir_bands.astype(np.float32)
red_bands = red_bands.astype(np.float32)
green_bands = green_bands.astype(np.float32)
blue_bands = blue_bands.astype(np.float32)
```

```python
# Stack the bands
stacked_image = np.stack([red_bands, green_bands, blue_bands], axis=-1)
```

```python

# Normalize pixel values to the range [0, 1]
stacked_image_normalized = stacked_image / stacked_image.max()
```

```python

# Plot the stacked image
plt.figure(figsize=(10, 10))
plt.imshow(stacked_image_normalized)
plt.title('Stacked Sentinel Image')
plt.axis('off')
plt.show()
```
